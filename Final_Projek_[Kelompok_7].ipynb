##**Statistical Modelling**



> Akan dilakukan pemodelan untuk memprediksi pengaruh luas panen, curah hujan, suhu rata-rata, dan kelembapan terhadap produksi padi dengan menggunakan pemodelan regression analysis



####**Import Library**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn import metrics
import joblib
from sklearn.preprocessing import OneHotEncoder

#Evaluasi Model
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import mean_absolute_percentage_error
```

####**Encoding Data**

Karena variabel provinsi memiliki data yang bertipe kategorikal, maka kita perlu melakukan encoding data.

```python
# Melakukan One Hot Encoding
onehot = OneHotEncoder()
df_onehot = onehot.fit_transform(df[['Provinsi']]).toarray()
df_onehot
```

```python
# Mengambil nama feature dari hasil encode
onehot.get_feature_names_out()
```

```python
# Meletakkan hasil encode ke pandas dataframe
df_encoded = df.copy()
df_encoded.drop(['Provinsi'], axis=1, inplace=True)
df_encoded[onehot.get_feature_names_out()] = df_onehot
df_encoded
```

```python
df_encoded = pd.get_dummies(df, columns=['Provinsi'])
df_encoded
```

####**Selection Variabel**

Kolom Encoding tiap provinsi, Luas Panen, Curah Hujan, Kelembaban, dan Suhu Rata-Rata akan menjadi variabel independen. Sementara kolom Prediksi akan menjadi variabel dependen.

```python
x = df_encoded.drop(['Produksi'], axis=1)
y = df_encoded[['Produksi']]
x.shape, y.shape
```

####**Split Dataset**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state = 70)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

####**Modelling**

##### **Linear Regression**

```python
# Initializing data
lr_model = LinearRegression()

# Fitting data
lr_model.fit(x_train, y_train)

# Menyimpan model ke file dengan ekstensi .pkl
joblib.dump(lr_model, 'model_regresi.pkl')

# Membaca kembali model dari file .pkl
loaded_model = joblib.load('model_regresi.pkl')

# Prediction
y_pred = loaded_model.predict(x_test)

# R2
test_R2_LR = r2_score(y_test, y_pred)

print('test R2 SCORE:', test_R2_LR)
```

**Interpretasi**


----
Output di atas merupakan nilai koefisien determinasi yang digunakan untuk mengukur seberapa baik model regresi linear sesuai dengan data yang diolah. Nilai koefisien determinasi berkisar antara 0 sampai 1. Apabila nilai koefisien determinasi mendekati 0 dapat diartikan bahwa model yang dibangun tidak dapat menjelaskan variasi sama sekali dan apabila nilai koefisien determinasi mendekati 1 maka dapat diartikan bahwa model yang dibentuk dapat menjelaskan seluruh variasi data yang ada.

Berdasarkan kasus di atas, didapatkan nilai koefisien determinasinya (R square) sebesar 0.895. Hal ini menunjukkan bahwa sekitar 89.5% dari variasi dalam variabel respon dapat dijelaskan oleh model. Sedangkan sisanya sekitar 10.5% dari variasi dalam variabel respon tidak dapat dijelaskan oleh model dan mungkin disebabkan oleh faktor-faktor lain yang tidak dimasukkan ke dalam model.

```python
lr_mae = mean_absolute_error(y_test, y_pred)
lr_mse = mean_squared_error(y_test, y_pred)
lr_mape = mean_absolute_percentage_error(y_test, y_pred)

print("Hasil MAE dari model Linear Regression:", lr_mae)
print("Hasil MSE dari model Linear Regression:", lr_mse)
print("Hasil MAPE dari model Linear Regression:", lr_mape)
```

##### **Random Forest Regressor**

```python
# Initializing data
rfReg_model = RandomForestRegressor()

# Fitting data
rfReg_model.fit(x_train, y_train.values.ravel())

# Menyimpan model ke file dengan ekstensi .pkl
joblib.dump(rfReg_model, 'model_RFRegressor.pkl')

# Membaca kembali model dari file .pkl
loaded_rfReg_model = joblib.load('model_RFRegressor.pkl')

# Prediction
y_pred = loaded_rfReg_model.predict(x_test)

# R2
test_R2_RF = r2_score(y_test.values.ravel(), y_pred)

print('test R2 SCORE:', test_R2_RF)
```

**Interpretasi**

R2 score mengukur sejauh mana variabilitas dari variabel dependen dapat dijelaskan oleh model regresi. Rentang nilai R2 score adalah 0 hingga 1, di mana nilai 1 menunjukkan bahwa model dapat menjelaskan seluruh variabilitas data, sedangkan nilai 0 berarti model tidak memberikan penjelasan sama sekali.

Hasil R2 score yang didapatkan untuk model Random Forest Regressor adalah sebesar 0.928 atau 92.8%, angka ini cukup tinggi, yang artinya variabel independen (luas panen, curah hujan, kelembapan, suhu rata-rata) berhasil dengan baik dalam menjelaskan variasi dalam data produksi padi.

```python
rfReg_mae = mean_absolute_error(y_test, y_pred)
rfReg_mse = mean_squared_error(y_test, y_pred)
rfReg_mape = mean_absolute_percentage_error(y_test, y_pred)

print("Hasil MAE dari model Random Forest Regressor:", rfReg_mae)
print("Hasil MSE dari model Random Forest Regressor:", rfReg_mse)
print("Hasil MAPE dari model Random Forest Regressor:", rfReg_mape)
```

##### **K-Nearest Neighbors Regressor**

K adalah parameter penting di dalam algoritma KNN, maka dari itu dibutuhkan pendekatan untuk mencari nilai K terbaik yg akan digunakan.

di bawah ini digunakan pendekatan menggunakan cross validation dengan metrik yg dipertimbangkan adalah MSE

```python
from sklearn.model_selection import cross_val_score

# nilai k yg diuji
k_values = range(1, 16)

# array utk menyimpan hasil skor evaluasi utk tiap k
cross_val_scores = []

for k in k_values:
    knn_regressor = KNeighborsRegressor(n_neighbors=k)
    scores = cross_val_score(knn_regressor, x_train, y_train, cv=5, scoring='neg_mean_squared_error')
    cross_val_scores.append(-scores.mean())

# mencari k terbaik
best_k = k_values[cross_val_scores.index(min(cross_val_scores))]

print("Nilai K terbaik:", best_k)
```

Di bawah ini juga akan dicoba pendekatan mencari nilai K dengan cara elbow method

```python
def Elbow(K):
  #initiating an empy list
  test_mse =[]

  #train model for every value of K
  for i in K:
    knnRegressor_model = KNeighborsRegressor(n_neighbors=i)
    knnRegressor_model.fit(x_train, y_train)
    tmp = knnRegressor_model.predict(x_test)
    tmp = mean_absolute_error(tmp, y_test)
    test_mse.append(tmp)

  return test_mse
```

```python
K = range(1, 16)
test = Elbow(K)

#plotting
plt.plot(K,test)
plt.xlabel('K Neighbors')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Elbow Curve for Test')
```

Dari grafik di atas, elbow terletak pada K=4, K=5, dan K=6, maka dari itu akan dilakukan percobaan KNNReg dengan ketiga nilai K dan akan dilihat nilai K yg menghasilkan R2 score paling baik



```python
# Inisialisasi data
k_values = [4, 5, 6]

# Loop untuk mencoba setiap nilai K
for k in k_values:
    # Inisialisasi model
    knnReg_model = KNeighborsRegressor(n_neighbors=k)

    # Fitting data
    knnReg_model.fit(x_train, y_train)

    # Prediction
    y_pred = knnReg_model.predict(x_test)

    # R2
    test_R2_KNN = r2_score(y_test, y_pred)

    # Print hasil R2 score
    print(f'R2 Score for K={k}: {test_R2_KNN}')
```

Dari ketiga nilai K, hasil R2 score paling tinggi didapatkan oleh K=6, maka dari itu model yg akan disimpan adalah KNNReg dengan K=6

```python
# Initializing data
knnReg_model = KNeighborsRegressor(n_neighbors=6)

# Fitting data
knnReg_model.fit(x_train, y_train.values.ravel())

# Menyimpan model ke file dengan ekstensi .pkl
joblib.dump(knnReg_model, 'model_KNNRegressor.pkl')

# Membaca kembali model dari file .pkl
loaded_knnReg_model = joblib.load('model_KNNRegressor.pkl')
```

R2 score mengukur sejauh mana variabilitas dari variabel dependen dapat dijelaskan oleh model regresi. Rentang nilai R2 score adalah 0 hingga 1, di mana nilai 1 menunjukkan bahwa model dapat menjelaskan seluruh variabilitas data, sedangkan nilai 0 berarti model tidak memberikan penjelasan sama sekali.

Hasil R2 score yang didapatkan untuk model KNN Regressor dengan k=6 adalah sebesar 0.841 atau 84.1%, angka ini cukup tinggi, yang artinya variabel independen (luas panen, curah hujan, kelembapan, suhu rata-rata) berhasil dengan baik dalam menjelaskan variasi dalam data produksi padi.
